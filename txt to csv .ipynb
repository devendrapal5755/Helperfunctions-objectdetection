```python
def convert_txt_to_csv(label_path):
    lisy = []
    for i in os.listdir(label_path):
        if '.txt' in i:
            with open(os.path.join(label_path,i)) as f:
                lines = f.readlines()
                for line in lines:
                    lisy.append(line.strip())
        if '.jpg' in i:
            name = pd.Series(data = i)
    series = pd.Series(data = lisy)
    df = pd.DataFrame(data = series)
    df = df[0].str.split(" ",expand = True)
    df = df.rename(columns = {0:'class',1:'x',2:'y',3:'h',4:'w'})
    df.insert(0, "img", df.min(axis=1))
    df['img'] = name
    df.to_csv('label.csv')
    return df
```
