```python
import glob 
import os 
import pandas as pd 
```

```python
os.chdir("/content/drive/MyDrive/k_dataset")
df = pd.DataFrame()
my_dict = {"file_d" : [], "lines_d" : []}
for file in glob.glob("*.txt"):
    with open(file) as f:
      lines = f.readlines()
      my_dict["file_d"].append(file)
      my_dict["lines_d"].append(lines) 
      # print(f"{file} data{lines}")
```

```python
df = pd.DataFrame.from_dict(my_dict)
```
